```python
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.pylab as pylab
pylab.rcParams['figure.figsize'] = 18, 8
plt.style.use('ggplot')
```

```python
df = pd.read_csv('euribor_12m_data.csv', delimiter =';', index_col='date')
df.dropna()
df.index = pd.to_datetime(df.index, format='%d/%m/%Y')
df['rate'] = pd.to_numeric(df['rate'], errors='coerce')
```

```python
month = df.rate.pct_change().groupby(df.index.month).mean()
```

```python
print(month)
```

```python
month.plot(kind='bar')
```

```python
media_rate = df['rate'].mean()

# Imprimir el resultado
print(media_rate)
```

```python
year = df.rate.pct_change().groupby(df.index.year).mean()
print(year)
year.plot(kind='bar')
```
